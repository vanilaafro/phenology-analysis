```python
import requests
from datetime import datetime

# Define parameters
region_code = "US-CA"  # Example region code, change as needed
species_code = "cangoo"  # Example species code (e.g., Canada Goose)
api_token = "your-token"  # Replace with your eBird API key

# Dynamically determine date range
current_year = datetime.today().year
start_year = current_year - 10  # Fetch data for the past 10 years
end_year = current_year

# Spring months (March to May)
spring_months = ["03", "04", "05"]
day = "01"  # Fixed day for consistency

# Base API URL (without species filtering)
api_base_url = "https://api.ebird.org/v2/data/obs/{}/{}/historic/{}/{}/{}/"

# Set request headers
headers = {"X-eBirdApiToken": api_token}

# Loop through years and spring months to fetch data
for year in range(start_year, end_year + 1):
    for month in spring_months:
        url = api_base_url.format(region_code, species_code, year, month, day)
        response = requests.get(url, headers=headers)

        if response.status_code == 200:
            data = response.json()

            # Filter for the specific species
            filtered_data = [entry for entry in data if entry.get("speciesCode") == species_code]

            if filtered_data:
                print(f"Data for {year}-{month}: {filtered_data}")
            else:
                print(f"No data for {species_code} in {year}-{month}")
        else:
            print(f"Failed to fetch data for {year}-{month}: {response.status_code}, {response.text}")
```

```python
import pandas as pd
```

```python
df = pd.DataFrame(data)
# Display DataFrame
df
```

```python
df["howMany"].sum()
#sum of sightings
```

```python
df_grouped = df.groupby('obsDt')['howMany'].sum().reset_index()
```

```python
import matplotlib.pyplot as plt

plt.figure(figsize=(12, 6))
plt.plot(df_grouped['obsDt'], df_grouped['howMany'], marker='o', linestyle='-')

plt.xlabel("Date")
plt.ylabel("Number of Observations (howMany)")
plt.title("Observations Over Time")
plt.xticks(rotation=45)  # Rotate x-axis labels for readability
plt.grid()
plt.show()
```

```python
# import matplotlib.pyplot as plt

# Sort the DataFrame by 'howMany' in descending order
df_sorted = df_grouped.sort_values(by='howMany', ascending=False)

plt.figure(figsize=(30, 6))
plt.bar(df_sorted['obsDt'], df_sorted['howMany'], color='skyblue')

plt.xlabel("Date")
plt.ylabel("Total Observations (Sum of howMany)")
plt.title("Total Observations Distribution (Sorted by Observation Count)")
plt.xticks(rotation=45)  # Rotate x-axis labels for readability
plt.grid(axis='y', linestyle='--', alpha=0.7)  # Grid for readability
plt.show()
```

```python
# import matplotlib.pyplot as plt

plt.figure(figsize=(12, 6))
plt.hist(df_grouped['howMany'], bins=20, color='skyblue', edgecolor='black', alpha=0.7)

plt.xlabel("Number of Observations (howMany)")
plt.ylabel("Frequency")
plt.title("Histogram of Observations Count Distribution")
plt.grid(axis='y', linestyle='--', alpha=0.7)  # Add grid for readability
plt.show()
```
